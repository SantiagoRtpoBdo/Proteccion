3.	Identificar si existe correlación entre las columnas “Edad”  y “Prepensión”; e imprimir el resultado.

No se puede hacer una correlacion entre “Edad” y “Prepensión”, ya que uno es entero y el otro objeto, no obstante si se realiza su devido proceso para poder poner “Prepensión” en un entero ya sea con Label Encoding o con One-Hot Encoding hallar la correlacion seria posible.

```python
df.corr()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Prepensión_encoded'] = label_encoder.fit_transform(df['Prepensión'])

correlation = df['Edad'].corr(df['Prepensión_encoded'])
print(f"Correlación entre Edad y Prepensión_encoded: {correlation}")
```

```python
df.corr()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

corr_matrix = df.corr()
fig, ax = plt.subplots(figsize = (8,8))
sn.heatmap(corr_matrix, annot=True)
plt.show()
```

4.	Identificar cuáles de los empleados cumplen la siguiente condición simultáneamente: NOTA EXAMEN 1 mayor o igual a 80, NOTA EXAMEN 2 mayor o igual a 75 y laboran en la ciudad de Bogotá. imprimir el resultado.

```python
# Filtrar los empleados que cumplen con la condición
condicion = (df['Nota 1'] >= 80) & (df['Nota 2'] >= 75) & (df['Ciudad'] == 'Bogotá')
empleados_cumplen_condicion = df[condicion]
print("Empleados que cumplen con la condición:")
print(empleados_cumplen_condicion)
```

```python
total_empleados_cumplen_condicion = len(empleados_cumplen_condicion)
print("Total de empleados que cumplen con la condición:", total_empleados_cumplen_condicion)
```
